```python
def intmethod(interestRate = 500, principal = 1000):
    nDays = 5
    interestsPct = ( 10000 + interestRate )**nDays - 10000**nDays
    print(interestsPct)
    return (principal*interestsPct)/10000**nDays

def normalmethod(interestRate = 5, principal = 1000):
    nDays = 5
    interestsPct = ( 1 + interestRate/100 )**nDays - 1
    print(interestsPct)
    return principal*interestsPct
            
print(intmethod())
print(normalmethod())
```
